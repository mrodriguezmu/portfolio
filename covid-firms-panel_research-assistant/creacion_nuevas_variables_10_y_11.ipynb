```python
# variable region
region = []
for i in datos['Ciudad']:
  if i == 'BOGOTA' or i == 'CHIA' or i == 'SOACHA' or i == 'CAJICA' or i == 'COTA' or i == 'MOSQUERA' or i == 'FUNZA' or i == 'TOCANCIPA' or i == 'FUSAGASUGA' or i == 'MADRID' or i == 'ZIPAQUIRA' or i == 'TENJO' or i == 'SIMIJACA' or i == 'CAQUEZA' or i == 'FACATATIVA' or i == 'COGUA' or i == 'UBATE' or i == 'SIBATE' or i == 'EL ROSAL' or i == 'GIRARDOT' or i == 'SUBACHOQUE':
    region.append('BOGOTA')
  elif  i == 'BUCARAMANGA' or i == 'FLORIDABLANCA' or i == 'GIRON' or i == 'CUCUTA' or i == 'BARBOSA' or i == 'TUNJA':
    region.append('NORORIENTE')
  elif  i == 'MEDELLIN' or i == 'ITAGUI' or i == 'SABANETA' or i == 'BELLO' or i == 'ENVIGADO' or i == 'LA ESTRELLA' or i == 'RIONEGRO' or i == 'GIRARDOTA' or i == 'COPACABANA' or i == 'ENTRERRIOS' or i == 'SANTA BARBARA' or i == 'URRAO' or i == 'CAREPA' or i == 'YARUMAL' or i == 'DON MATIAS':
    region.append('NOROCCIDENTE')
  elif  i == 'ACACIAS' or i == 'GRANADA' or i == 'RESTREPO' or i == 'GUAMAL' or i == 'LETICIA':
    region.append('SURORIENTE')
  elif  i == 'NEIVA' or i == 'IBAGUE' or i == 'CAMPOALEGRE' or i == 'FLORENCIA':
    region.append('CENTRO')
  elif  i == 'MANIZALES' or i == 'PEREIRA' or i == 'ARMENIA' or i == 'ANSERMA' or i == 'SANTUARIO' or i == 'DOSQUEBRADAS':
    region.append('EJE CAFETERO')
  elif  i == 'CALI' or i == 'YUMBO' or i == 'PALMIRA' or i == 'TULUA' or i == 'CARTAGO' or i == 'CANDELARIA' or i == 'SAN PEDRO' or i == 'BUGA' or i == 'SANTANDER DE QUILICHAO' or i == 'PASTO' or i == 'LA UNION' or i == 'JAMUNDI' or i == 'POPAYAN' or i == 'ANSERMANUEVO' or i == 'BUENAVENTURA' or i == 'EL TAMBO' or i == 'SEVILLA' or i == 'BUGALAGRANDE' or i == 'PUERTO TEJADA' or i == 'CAJIBIO' or i == 'TANGUA' or i == 'ZARZAL' or i == 'ROLDANILLO' or i == 'PRADERA' or i == 'CALOTO':
    region.append('SUROCCIDENTE')
  else:
    region.append('COSTA CARIBE')
datos['Región'] = region
```
