```python
# variable region
region = []
for i in datos['Estado/Provincia/Región']:
  if i == 'Bogotá D.C' or i == 'Cundinamarca' or i == 'Bogota D.C':
    region.append('Bogotá')
  elif  i == 'Arauca' or i == 'Boyacá' or i == 'Norte de Santander' or i == 'Santander' or i == 'Boyaca':
    region.append('Nororiente')
  elif  i == 'Antioquia' or i == 'Chocó' or i == 'Choco':
    region.append('Noroccidente')
  elif  i == 'Amazonas' or i == 'Casanare' or i == 'Guainía' or i == 'Guainia' or i == 'Guaviare' or i == 'Meta' or i == 'Vaupés' or i == 'Vaupes' or i == 'Vichada':
    region.append('Suroriente')
  elif  i == 'Caquetá' or i == 'Huila' or i == 'Tolima' or i == 'Caqueta':
    region.append('Centro')
  elif  i == 'Caldas' or i == 'Quindío' or i == 'Risaralda' or i == 'Quindio':
    region.append('Eje Cafetero')
  elif  i == 'Cauca' or i == 'Nariño' or i == 'Putumayo' or i == 'Valle del Cauca':
    region.append('Suroccidente')
  else:
    region.append('Costa Caribe')
datos['Región'] = region
```

```python
# Creación de variables binarias que indica a qué sector pertenecen (10 u 11)
# 10
cod_10 = []
for i in datos['Sector (CIIU Rev.4 A.C.)']:
  if 'C10' in i:
    cod_10.append(1) # pertenece al codigo 10
  else:
    cod_10.append(0) # no pertenece al codigo 10
```

```python
# 11
cod_11 = []
for i in datos['Sector (CIIU Rev.4 A.C.)']:
  if 'C11' in i:
    cod_11.append(1) # pertenece al codigo 11
  else:
    cod_11.append(0) # no pertenece al codigo 11
```

```python
datos['Código 10'] = cod_10
```

```python
datos['Código 11'] = cod_11
```

```python
# Crear variable que indique el sector especifico
especi = []
for i in datos['Sector (CIIU Rev.4 A.C.)']:
  division = i.split('; ')
  cada_uno = []
  for a in division:
    if 'C10' in a or 'C11' in a:
      cada_uno.append(a)
  especi.append(cada_uno)
```

```python
datos['Sector Específico'] = especi
```

```python
# Crear variable binaria para saber si una empresa exporta o no
exporta = []
for i in datos['Exportar']:
  if i != '.':
    exporta.append('si') # valor de 1 para los exportadores
  else:
    exporta.append('no') # 0 para los no exportadores
```

```python
datos['Exportación'] = exporta
```

```python
# Crear variable binaria para saber si una empresa importa o no
importa = []
for i in datos['Importar']:
  if i != '.':
    importa.append('si') # valor de 1 para los exportadores
  else:
    importa.append('no') # 0 para los no exportadores
```

```python
datos['Importación'] = importa
```

```python
# Creación variable riesgo a partir del ROA
riesgo1=[]
riesgo2=[]
riesgo3=[]
riesgo4=[]
for i in datos['Rendimiento Sobre Los Activos (ROA) (%)']:
  if i <= 0.68:
    riesgo1.append(1)
    riesgo2.append(0)
    riesgo3.append(0)
    riesgo4.append(0)
  elif 0.68 < i <= 3.42:
    riesgo1.append(0)
    riesgo2.append(1)
    riesgo3.append(0)
    riesgo4.append(0)
  elif 3.42 < i <= 7.6125:
    riesgo1.append(0)
    riesgo2.append(0)
    riesgo3.append(1)
    riesgo4.append(0)
  else:
    riesgo1.append(0)
    riesgo2.append(0)
    riesgo3.append(0)
    riesgo4.append(1)
```

```python
datos['R_1'] = riesgo1
datos['R_2'] = riesgo2
datos['R_3'] = riesgo3
datos['R_4'] = riesgo4
```
