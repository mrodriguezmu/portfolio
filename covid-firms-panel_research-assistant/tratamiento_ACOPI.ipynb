```python
# Unificación representante legal
repre=[]
indice=0
for i in datos['5. Nombre del representante legal']:
  if datos.loc[indice, '5. Nombre del representante legal'] == 'n.d.' and datos.loc[indice, '5. Nombre del representante legal.1'] == 'n.d.':
    repre.append('n.d')
  elif datos.loc[indice, '5. Nombre del representante legal'] == 'n.d.' and datos.loc[indice, '5. Nombre del representante legal.1'] != 'n.d.':
    repre.append(datos.loc[indice, '5. Nombre del representante legal.1'])
  elif datos.loc[indice, '5. Nombre del representante legal'] != 'n.d.' and datos.loc[indice, '5. Nombre del representante legal.1'] == 'n.d.':
    repre.append(datos.loc[indice, '5. Nombre del representante legal'])
  else:
    repre.append(datos.loc[indice, '5. Nombre del representante legal'])
  indice = indice +1
```

```python
datos['Representante legal']=repre
```

```python
# Unificación sexo del representante
sexo=[]
indice=0
for i in datos['6. Indique el sexo del representante legal ']:
  if datos.loc[indice, '6. Indique el sexo del representante legal '] == 'n.d.' and datos.loc[indice, '6. Indique el sexo del representante legal .1'] == 'n.d.':
    sexo.append('n.d')
  elif datos.loc[indice, '6. Indique el sexo del representante legal '] == 'n.d.' and datos.loc[indice, '6. Indique el sexo del representante legal .1'] != 'n.d.':
    sexo.append(datos.loc[indice, '6. Indique el sexo del representante legal .1'])
  elif datos.loc[indice, '6. Indique el sexo del representante legal '] != 'n.d.' and datos.loc[indice, '6. Indique el sexo del representante legal .1'] == 'n.d.':
    sexo.append(datos.loc[indice, '6. Indique el sexo del representante legal '])
  else:
    sexo.append(datos.loc[indice, '6. Indique el sexo del representante legal '])
  indice = indice +1
```

```python
rango = range(len(datos.index))
```

```python
# ROA con P/G operacionales [=EBIT]
roao = []
for indice in rango:
  try:
    roa = float(datos.loc[indice, 'P/G operacionales [=EBIT]'])/float(datos.loc[indice, 'Activos totales'])
    roao.append(roa)
  except:
    roao.append('n.d.')
```

```python
len(roao)
```

```python
datos['ROA con P/G operacionales [=EBIT]'] = roao
```

```python
# ROA con Result. ordinarios antes impuestos
roar = []
for indice in rango:
  try:
    roa = float(datos.loc[indice, 'Result. ordinarios antes impuestos'])/float(datos.loc[indice, 'Activos totales'])
    roar.append(roa)
  except:
    roar.append('n.d.')
```

```python
datos['ROA con Result. ordinarios antes impuestos'] = roar
```

```python
# ROA con P/G por periodo [=ingresos netos]
roan = []
for indice in rango:
  try:
    roa = float(datos.loc[indice, 'P/G por periodo [=ingresos netos]'])/float(datos.loc[indice, 'Activos totales'])
    roan.append(roa)
  except:
    roan.append('n.d.')
```

```python
datos['ROA con P/G por periodo [=ingresos netos]'] = roan
```

```python
patri=[]
for indice in rango:
  if datos.loc[indice, 'Fondos de los accionistas'] != 'n.d.':
    if datos.loc[indice, 'Capital'] != 'n.d.' and datos.loc[indice, 'Otros accionistas de fondos'] == 'n.d.':
      patrimo = float(datos.loc[indice, 'Fondos de los accionistas']) + float(datos.loc[indice, 'Capital'])
      patri.append(patrimo)
    elif datos.loc[indice, 'Capital'] == 'n.d.' and datos.loc[indice, 'Otros accionistas de fondos'] != 'n.d.':
      patrimo = float(datos.loc[indice, 'Fondos de los accionistas']) + float(datos.loc[indice, 'Otros accionistas de fondos'])
      patri.append(patrimo)
    elif datos.loc[indice, 'Capital'] != 'n.d.' and datos.loc[indice, 'Otros accionistas de fondos'] != 'n.d.':
      patrimo = float(datos.loc[indice, 'Fondos de los accionistas']) + float(datos.loc[indice, 'Otros accionistas de fondos'])+float(datos.loc[indice, 'Capital'])
      patri.append(patrimo)
    else:
      patrimo = float(datos.loc[indice, 'Fondos de los accionistas'])
      patri.append(patrimo)
  else:
    patri.append('n.d.')
```

```python
datos['Patrimonio total'] = patri
```

```python
pas =[]
for indice in rango:
  try:
    pasi = float(datos.loc[indice, 'Activos totales']) - float(datos.loc[indice, 'Patrimonio total'])
    pas.append(pasi)
  except:
    pas.append('n.d.')
```

```python
datos['Pasivos totales'] = pas
```

```python
pasc = []
for indice in rango:
  try:
    pasic = float(datos.loc[indice, 'Pasivos totales']) - float(datos.loc[indice, 'Pasivos no corrientes'])
    pasc.append(pasic)
  except:
    pasc.append('n.d.')
```

```python
datos['Pasivos corrientes'] = pasc
```

```python
intensidadt = []
for indice in rango:
  try:
    intensi = float(datos.loc[indice, 'Activos totales']) / float(datos.loc[indice, 'Ingresos explotación (Cifra ventas)'])
    intensidadt.append(intensi)
  except:
    intensidadt.append('n.d.')
```

```python
datos['Intensidad tecnológica'] = intensidadt
```

```python
holfi = []
for indice in rango:
  try:
    holgufi = float(datos.loc[indice, 'Activo circulante']) / float(datos.loc[indice, 'Pasivos corrientes'])
    holfi.append(holgufi)
  except:
    holfi.append('n.d.')
```

```python
datos['Holgura financiera'] = holfi
```

```python
apalanca = []
for indice in rango:
  try:
    miento = float(datos.loc[indice, 'Pasivos totales']) / float(datos.loc[indice, 'Activos totales'])
    apalanca.append(miento)
  except:
    apalanca.append('n.d.')
```

```python
datos['Apalancamiento']=apalanca
```

```python
# presion financiera sin usar depreciacion
pres=[]
for indice in rango:
  if datos.loc[indice, 'Gastos financieros'] != 'n.d.' and datos.loc[indice, 'Result. ordinarios antes impuestos'] != 'n.d.':
    presion = float(datos.loc[indice, 'Gastos financieros'])/(float(datos.loc[indice, 'Result. ordinarios antes impuestos'])+float(datos.loc[indice, 'Gastos financieros']))
    pres.append(presion)
  else:
    pres.append('n.d.')
```

```python
datos['Presión financiera'] = pres
```

```python
sol = []
for indice in rango:
  try:
    vencia = float(datos.loc[indice, 'Activos totales']) / float(datos.loc[indice, 'Pasivos totales'])
    sol.append(vencia)
  except:
    sol.append('n.d.')
```

```python
datos['Solvencia'] = sol
```

```python
holp1 = []
for indice in rango:
  try:
    holgura = float(datos.loc[indice, 'Efectivo y equivalentes de efectivo'])/float(datos.loc[indice, 'Deudas a corto plazo'])
    holp1.append(holgura)
  except:
    holp1.append('n.d.')
```

```python
datos['Holgura potencial (con Deudas a corto plazo'] = holp1
```

```python
holp2 = []
for indice in rango:
  try:
    holgura = float(datos.loc[indice, 'Efectivo y equivalentes de efectivo'])/float(datos.loc[indice, 'Pasivos corrientes'])
    holp2.append(holgura)
  except:
    holp2.append('n.d.')
```

```python
datos['Holgura potencial (con Pasivos corrientes'] = holp2
```

```python
# cash flow sin amortizaciones (porque no se encuentran en la base de datos), y teniendo en cuenta que la variable provisiones
# no está disponible para la mayoria de observaciones.
# recordar que estandarizar por activos totales significa expresar la variable en términos de porcentaje de los activos totales

cash = []
for indice in rango:
  if datos.loc[indice, 'Provisiones'] == 'n.d.':
    try:
      flow = float(datos.loc[indice, 'P/G por periodo [=ingresos netos]'])/float(datos.loc[indice, 'Activos totales'])
      cash.append(flow)
    except:
      cash.append('n.d.')
  else:
    try:
      flow = (float(datos.loc[indice, 'P/G por periodo [=ingresos netos]'])+
              float(datos.loc[indice, 'Provisiones']))/float(datos.loc[indice, 'Activos totales'])
      cash.append(flow)
    except:
      cash.append('n.d.')
```

```python
datos['Cash flow'] = cash
```

```python
rota = []
for indice in rango:
  try:
    cion = float(datos.loc[indice, 'Ingresos explotación (Cifra ventas)'])/float(datos.loc[indice, 'Activos totales'])
    rota.append(cion)
  except:
    rota.append('n.d.')
```

```python
datos['Rotación_de_activos'] = rota
```

```python
rapat = []
for indice in rango:
  try:
    act = float(datos.loc[indice, 'Patrimonio total'])/float(datos.loc[indice, 'Activos totales'])
    rapat.append(act)
  except:
    rapat.append('n.d.')
```

```python
datos['Razón_de_patrimonio_a_activo_total'] = rapat
```

```python
comp = []
for indice in rango:
  try:
    deu = float(datos.loc[indice, 'Pasivos corrientes'])/float(datos.loc[indice, 'Pasivos totales'])
    comp.append(deu)
  except:
    comp.append('n.d.')
```

```python
datos['Composición_de_la_deuda'] = comp
```

```python
rent = []
for indice in rango:
  try:
    op = float(datos.loc[indice, 'P/G por periodo [=ingresos netos]'])/float(datos.loc[indice, 'Ingresos explotación (Cifra ventas)'])
    rent.append(op)
  except:
    rent.append('n.d.')
```

```python
datos['Rentabilidad_sobre_ingresos_operacionales'] = rent
```

```python
renta = []
for indice in rango:
  try:
    pat = float(datos.loc[indice, 'P/G por periodo [=ingresos netos]'])/float(datos.loc[indice, 'Patrimonio total'])
    renta.append(pat)
  except:
    renta.append('n.d.')
```

```python
datos['Rentabilidad_sobre_patrimoio'] = renta
```
