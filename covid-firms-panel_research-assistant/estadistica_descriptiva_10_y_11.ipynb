```python
# Creamos una nuevo Dataframe que mostrará cuantas variables en blanco tiene cada empresa
valores_en_blanco_pre = pd.DataFrame()
```

```python
empresas = []
año = []
blancos = []
rango = range(len(datos.index))
for indice in rango:
  contador = 0
  empresas.append(datos.loc[indice, "Nombre Empresa"])
  año.append(datos.loc[indice, "Año"])
  for columna in datos:
    if datos.loc[indice, columna] == 'n.d.' or datos.loc[indice, columna] == 'n.s.' or pd.isnull(datos.loc[indice, columna]) == True:
      contador = contador +1
  blancos.append(contador)
valores_en_blanco_pre["Nombre Empresa"] = empresas
valores_en_blanco_pre["Año"] = año
valores_en_blanco_pre["Variables en blanco"] = blancos
```

```python
valores_en_blanco_pre.tail(12)
```

```python
#sumamos el total de valores en blanco para cada empresa
nit_0 = valores_en_blanco_pre.loc[0, "Nombre Empresa"]
nuevo_ceros=[]
suma = 0
for indice in rango:
  nit_1 = valores_en_blanco_pre.loc[indice, "Nombre Empresa"]
  if indice != len(datos.index)-1:
    if nit_1 == nit_0:
      suma = suma + valores_en_blanco_pre.loc[indice, "Variables en blanco"]
      nit_0 = valores_en_blanco_pre.loc[indice, "Nombre Empresa"]
    else:
      nuevo_ceros.append(suma)
      suma = valores_en_blanco_pre.loc[indice, "Variables en blanco"]
      nit_0 = valores_en_blanco_pre.loc[indice, "Nombre Empresa"]
  else:
    suma = suma + valores_en_blanco_pre.loc[indice, "Variables en blanco"]
    nuevo_ceros.append(suma)
```

```python
valores_en_blanco = pd.DataFrame()
```

```python
valores_en_blanco["Nombre Empresa"]=list(set(valores_en_blanco_pre["Nombre Empresa"]))
```

```python
valores_en_blanco["Total"]=nuevo_ceros
```
