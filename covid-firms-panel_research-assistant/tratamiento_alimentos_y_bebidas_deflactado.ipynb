```python
datos = pd.read_excel('panel deflactado depurado.xlsx')
datos.info()
```

```python
rango = range(len(datos.index))
```

```python
# Eliminar aquellas empresas que tienen 0 en Costo de mercancias vendidas para al menos un año
nombres_cost = []
for indice in rango:
  if datos.loc[indice, 'Costo de mercancías vendidas (deflactado)'] == 0:
    nombres_cost.append(datos.loc[indice, 'CO-NIT'])
```

```python
a_eliminar_cost=list(set(nombres_cost))
```

```python
indice = 0
for nit in datos['CO-NIT']:
  if nit in a_eliminar_cost:
    datos = datos.drop(index = indice)
  indice = indice +1
```

```python
# Contar cuantas empresas hay por sector
sector = datos['Sector Específico'].tolist()
sector_edit = []
for i in sector:
  division = i.split('; ')
  for a in division:
    sector_edit.append(a)
sector_norep = list(set(sector_edit))
aparicio = []
for a in sector_norep:
  num = sector_edit.count(a)
  aparicio.append([a, num])
```
