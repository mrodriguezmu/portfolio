```python
# Importamos las distintas librerias que vamos a utilizar:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import stats
import scipy
import statsmodels.api as sm
import statsmodels.formula.api as smf
import plotly
%matplotlib inline
from sklearn.metrics import r2_score
```

```python
# lo primero es eliminar aquellas empresas cuyo ultimo año registrado no sea 2021, por lo que haré una funcion que haga justamente eso
datos = pd.read_excel('Export 16_12_2022 05_53 editado.xlsx')
datos.info()
```

```python
indice = 0
for año in datos['Último año disponible']:
  if año != 2021:
    datos = datos.drop(index = indice)
  indice = indice +1
```

```python
indice = 0
for codigo in datos['NACE Rev. 2 Código principal (4 digitos)']:
  if codigo < 1000 or codigo > 1199:
    datos = datos.drop(index = indice)
  indice = indice +1
```

```python
for columna in datos:
  if columna == 'Unnamed: 0':
    datos = datos.drop([columna], axis = 1)
```

```python
for columna in datos:
  datos.rename(columns={columna:columna.replace('mil USD Últ. año disp.', 'USD 2021')}, inplace = True)
  datos.rename(columns={columna:columna.replace('mil LCU Últ. año disp.', '2021')}, inplace = True)
  datos.rename(columns={columna:columna.replace('Últ. año disp.', '2021')}, inplace = True)
  datos.rename(columns={columna:columna.replace('mil LCU Año - 1', '2020')}, inplace = True)
  datos.rename(columns={columna:columna.replace('mil LCU Año - 2', '2019')}, inplace = True)
  datos.rename(columns={columna:columna.replace('mil LCU Año - 3', '2018')}, inplace = True)
  datos.rename(columns={columna:columna.replace('mil LCU Año - 4', '2017')}, inplace = True)
  datos.rename(columns={columna:columna.replace('mil LCU Año - 5', '2016')}, inplace = True)
```

```python
panel = pd.DataFrame()
```

```python
nombre = []
Inactive = []
Quoted = []
Branch = []
OwnData=[]
Woco=[]
PaísISOCódigo=[]
NACERevCódigoprincipal = []
Códigodeconsolidación = []
año_lista = []
Ingresosexplotación=[]
empleados2021=[]
ciudad = []
numeroidentidad=[]
roebase = []
rocebase=[]
roare=[]
roere=[]
rocere=[]
roain=[]
marbe=[]
marbru=[]
marebi=[]
ebit=[]
flujo=[]
valor=[]
market=[]
ratio=[]
presta=[]
netdep=[]
nettot=[]
asdep=[]
astot=[]
```

```python
#Solo aplica para las variables que no varian con los años
def nose(columna, lista):
  indice = 0
  for a in datos[columna]:
      lista.append(datos.loc[indice, columna])
      lista.append(datos.loc[indice, columna])
      lista.append(datos.loc[indice, columna])
      lista.append(datos.loc[indice, columna])
      lista.append(datos.loc[indice, columna])
      lista.append(datos.loc[indice, columna])
      indice = indice +1
```

```python
# Para crear la columna año
def años(filas):
  for a in range(filas):
    año_lista.append(2021)
    año_lista.append(2020)
    año_lista.append(2019)
    año_lista.append(2018)
    año_lista.append(2017)
    año_lista.append(2016)
```
