```python
# Con esta funcion puedo ver qué empresas y en que año reportan un valor de 0 para una variable especifica
def valores_raros(columna):
  resultado = []
  indice = 0
  for val in datos[columna]:
    if val == 0:
      adjuntar = [datos.loc[indice, "NIT"], datos.loc[indice, "Fecha de Corte"], indice]
      resultado.append(adjuntar)
    indice = indice + 1
  return resultado
```

```python
# Funcion para determinar qué empresas no cuentan con la variable edad
def edad_na():
  resultado = []
  indice = 0
  for val in panel["Edad (años)"]:
    if pd.isnull(val) == True:
      adjuntar = panel.loc[indice, "NIT"]
      resultado.append(adjuntar)
    indice = indice + 1
  return list(set(resultado))
```

```python
# Creamos una nuevo Dataframe que mostrará cuantas variables en 0 tiene cada empresa
valores_en_cero = pd.DataFrame()
```

```python
empresas = []
año = []
ceros = []
rango = range(5100)
for indice in rango:
  contador = 0
  empresas.append(panel.loc[indice, "NIT"])
  año.append(panel.loc[indice, "Fecha de Corte"])
  for columna in panel:
    if panel.loc[indice, columna] == 0:
      contador = contador +1
  ceros.append(contador)
valores_en_cero["NIT"] = empresas
valores_en_cero["Fecha de Corte"] = año
valores_en_cero["Variables en Cero"] = ceros
```

```python
valores_en_cero.head(10)
```

```python
valores_en_cero = valores_en_cero.drop("Fecha de Corte", axis=1)
```

```python
#sumamos el total de valores en cero para cada empresa
nit_0 = valores_en_cero.loc[0, "NIT"]
nuevo_ceros=[]
suma = 0
for indice in rango:
  nit_1 = valores_en_cero.loc[indice, "NIT"]
  if indice != 5099:
    if nit_1 == nit_0:
      suma = suma + valores_en_cero.loc[indice, "Variables en Cero"]
      nit_0 = valores_en_cero.loc[indice, "NIT"]
    else:
      nuevo_ceros.append(suma)
      suma = valores_en_cero.loc[indice, "Variables en Cero"]
      nit_0 = valores_en_cero.loc[indice, "NIT"]
  else:
    suma = suma + valores_en_cero.loc[indice, "Variables en Cero"]
    nuevo_ceros.append(suma)
```
